```python
import numpy as np
import matplotlib.pyplot as plt

N = 20
x = np.random.rand(N)
y = 5 + 2*x + 1e-1*np.random.randn(N)

X = np.hstack([np.ones((N,1)), x.reshape(N,1)])

theta = np.array([5, 2])
X_plot = np.hstack([np.ones((10,1)), np.linspace(0, 1, 10).reshape(-1,1)])
y_plot = X_plot.dot(theta)

plt.figure()
plt.scatter(x,y)
plt.plot(X_plot[:,1],y_plot)
```

```python
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
print(theta)

y_predict = X.dot(theta)
y_plot = X_plot.dot(theta)

SSE = np.sum( (y - y_predict)**2 )
print('SSE: %8.3f' % SSE)

plt.figure()
plt.scatter(x,y)
plt.plot(X_plot[:,1],y_plot)
```

```python
eta = 0.01
maxIter = 20

theta = np.random.randn(2)

y_predict = X.dot(theta)
y_plot = X_plot.dot(theta)

SSE = [ np.sum( (y - y_predict)**2 ) ]
print('SSE: %8.3f' % SSE[-1])

plt.figure()
plt.scatter(x,y)
plt.plot(X_plot[:,1],y_plot)

for iter in range(maxIter):
    grad = 2*X.T.dot(y_predict - y)
    theta = theta - eta*grad
    
    y_predict = X.dot(theta)
    y_plot = X_plot.dot(theta)

    SSE.append( np.sum( (y - y_predict)**2 ) )
    print('SSE: %8.3f' % SSE[-1])
    
    plt.plot(X_plot[:,1],y_plot)


plt.figure()
plt.plot(SSE)
```

```python
eta = 0.1
maxIter = 20

theta = np.random.randn(2)

y_predict = X.dot(theta)
y_plot = X_plot.dot(theta)

SSE = [ np.sum( (y - y_predict)**2 ) ]
print('SSE: %8.3f' % SSE[-1])

plt.figure()
plt.scatter(x,y)
plt.plot(X_plot[:,1],y_plot)

for iter in range(maxIter):
    idn = np.random.randint(0, N, 1)
    grad = 2*X[idn,:].reshape(2) * (y_predict[idn] - y[idn])
    theta = theta - eta*grad
    
    y_predict = X.dot(theta)
    y_plot = X_plot.dot(theta)

    SSE.append( np.sum( (y - y_predict)**2 ) )
    print('SSE: %8.3f' % SSE[-1])
    
    plt.plot(X_plot[:,1],y_plot)


plt.figure()
plt.plot(SSE)
```
